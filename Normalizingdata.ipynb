```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Create some LogNormal data 

```python
#create the data using Numpy Lognormal data
np.random.seed(43)
data = np.random.lognormal(0.7,1.5,1000)
mean = np.mean(data)
std = np.std(data)
usl = 100

#Calculate the CPk
lsl = 0
cpu = (usl - mean) / (3 * std)
cpl = (mean - lsl) / (3 * std)
cpk = min(cpu,cpl)

sig3 = (mean + (3*std))
sig3b = (mean - (3*std))

#start drawing graph
sns.set_theme('paper')
plt.figure(figsize=(10,8))
sns.histplot(data,bins=50,kde=True)
plt.axvline(sig3,color='g',linestyle='--')
plt.axvline(sig3b,color='g',linestyle='--')
plt.axvline(usl,color='r',linestyle='--')
plt.axvline(lsl,color='r',linestyle='--')
plt.axvline(mean,color='black',linestyle='--')
plt.text(mean+1,68,'Mean',rotation=90)
plt.text(usl+1,65,'USL = 100',rotation=90)
plt.text(sig3+1,65,'3 Sigma',rotation=90)
plt.text(lsl+1,65,'LSL = 0',rotation=90)

plt.title(f'Transforming Lognormal Data std = {std}')

print(f'Results \n Cpl = {round(cpl,2)} \n Cpu = {round(cpu,2)} \n CPk = {round(cpk,2)}')
```

## Reflection data transformation, Limits based on a pivet point. 
* Pivet point normally the max Value of the data.
* min/max specifications are based on the same scale.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate left-skewed data using Beta distribution
np.random.seed(42)
data = np.random.beta(2, 5, size=1000) * 100  # Scale to [0, 100]
USL = 90  # Upper Specification Limit
LSL = 30  # Lower Specification Limit

# Step 1: Reflection around the maximum value
pivot = max(data)
reflected_data = pivot - data

# Step 2: Adjust specification limits
USL_new = pivot - LSL
LSL_new = pivot - USL

# Step 3: Calculate statistics for reflected data
mean_reflected = np.mean(reflected_data)
std_dev_reflected = np.std(reflected_data)

# Step 4: Calculate CpK for the reflected data
cpk_upper = (USL_new - mean_reflected) / (3 * std_dev_reflected)
cpk_lower = (mean_reflected - LSL_new) / (3 * std_dev_reflected)
cpk = min(cpk_upper, cpk_lower)

# Visualize the original and reflected data
plt.figure(figsize=(12, 6))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(USL, color='red', linestyle='--', label='USL')
plt.axvline(LSL, color='green', linestyle='--', label='LSL')
plt.title("Original Left-Skewed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# Reflected data
plt.subplot(1, 2, 2)
plt.hist(reflected_data, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
plt.axvline(USL_new, color='red', linestyle='--', label='Adjusted USL')
plt.axvline(LSL_new, color='green', linestyle='--', label='Adjusted LSL')
plt.title("Reflected Right-Skewed Data")
plt.xlabel("Reflected Value")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

# Print results
print(f"Original Data Mean (): {np.mean(data):.2f}, Std Dev (): {np.std(data):.2f}")
print(f"Reflected Data Mean (): {mean_reflected:.2f}, Std Dev (): {std_dev_reflected:.2f}")
print(f"CpK (Upper): {cpk_upper:.2f}")
print(f"CpK (Lower): {cpk_lower:.2f}")
print(f"Final CpK: {cpk:.2f}")


```
