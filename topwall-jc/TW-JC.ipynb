```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Adding a title example

```python
n1 = pd.read_csv('TWMW-JC.csv',encoding='latin1',parse_dates=['Data'])

Indx = (['Produto', 'Linha', 'Turno', 'Data', 'Hora', 'Funcionario', 'Trimmers',
       'Processo', 'Tipo de Teste', 'Método', 'Instrumento',
       ' Media Parede Grossa ', ' Range Parede Grossa ',
       ' Max. Parede Grossa ', ' Min. Parede Grossa ', ' Média Parede Fina ',
       ' Range Parede Fina ', ' Max. Parede Fina ', ' Min. Parede Fina '])

df1 = pd.DataFrame(n1,columns=Indx)
print(df1.shape)
df1.head()
```

```python
df = df1[[' Media Parede Grossa ',' Média Parede Fina ']]
df.describe()
```

```python
df = df[df[' Media Parede Grossa '] < 0.006]
lsl = 0.0055
usl = 0.0059
target = 0.0057
med = np.mean(df[' Media Parede Grossa '])
sns.scatterplot(df,x=' Media Parede Grossa ',y=df.index)
plt.axvline(x=lsl,color='r')
plt.axvline(x=usl,color='r')
plt.axvline(x=target,color='r')
plt.axvline(x=med,color='g')
```

```python
df = df[df[' Média Parede Fina '] < 0.0040]
lsl = 0.0033
usl = 0.0037
target = 0.0035
med = np.mean(df[' Média Parede Fina '])
sns.scatterplot(df,x=' Média Parede Fina ',y=df.index)
plt.axvline(x=lsl,color='r')
plt.axvline(x=usl,color='r')
plt.axvline(x=target,color='r')
plt.axvline(x=med,color='g')
```
